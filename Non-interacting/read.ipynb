```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)

plt.rcParams.update({'font.size': 15})
plt.rcParams.update({'text.usetex': True})
plt.rcParams.update({'text.latex.preamble' : r'\usepackage{amsmath}'})
```

```python
data = np.loadtxt('baresus128n435.dat')
kx = np.linspace(-np.pi, np.pi, len(data[0]))
ky = np.linspace(-np.pi, np.pi, len(data[0]))
```

```python
A, B = np.meshgrid(kx, ky) 
X = data
# using automatic selection of 
# contour levels; 
figure1, axes2 = plt.subplots(constrained_layout = True) 
#C = axes2.contourf(A, B, X, 10, cmap = plt.cm.bwr, vmin=1, vmax=np.max(data)) 
C = axes2.contourf(A, B, X, cmap = plt.cm.bwr) 
 
#C2 = axes2.contour(C, levels = C.levels[::2],  
#                   colors ='r') 

#axes2.set_title(r'NGRID=128x128')
axes2.set_xlabel(r'$q_x$') 
axes2.set_ylabel(r'$q_y$') 

# Make a colorbar for the ContourSet 
# returned by the contourf call. 
cbar = figure1.colorbar(C) 
cbar.ax.set_ylabel(r'$\chi_0$') 

axes2.set_yticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r"$\pi/2$", r'$0$', r'$\pi/2$', r'$\pi$'])
axes2.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r"$\pi/2$", r'$0$', r'$\pi/2$', r'$\pi$'])

#plt.savefig('colormap_n435.jpg')
plt.savefig('colormap_n435.pdf')
plt.show() 
```

```python
A, B = np.meshgrid(kx, ky) 
X = data
fig, ax = plt.subplots()
C = ax.pcolormesh(A, B, X, cmap='seismic')
fig.colorbar(C)
plt.xlim(-np.pi, np.pi)
plt.ylim(-np.pi, np.pi)
plt.savefig('test-comp.pdf')
```
