```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 15})
plt.rcParams.update({'text.usetex': True})
plt.rcParams.update({'text.latex.preamble' : r'\usepackage{amsmath}'})
```

```python
orb1, orb2, orb3, orb4, orb5 = [], [] ,[], [], []
rho, w = [], []
```

```python
r1 = np.loadtxt('R1.txt')
r2 = np.loadtxt('R2.txt')
r3 = np.loadtxt('R3.txt')

orb1 = np.append(r1[:,0], [r2[:,0], np.flip(r3[:,0])])
orb2 = np.append(r1[:,1], [r2[:,1], np.flip(r3[:,1])])
orb3 = np.append(r1[:,2], [r2[:,2], np.flip(r3[:,2])])
orb4 = np.append(r1[:,3], [r2[:,3], np.flip(r3[:,3])])
orb5 = np.append(r1[:,4], [r2[:,4], np.flip(r3[:,4])])
```

```python
size1 = len(r1[:,0])
size2 = size1+len(r2[:,0])
size3 = size2+len(r3[:,0])
sizeT = len(orb1)

num = np.linspace(0, sizeT, size3)
```

```python
print(size3)
```

```python
plt.plot(num, orb1, color='blue')
plt.plot(num, orb2, color='orangered')
plt.plot(num, orb3, color='gold')
plt.plot(num, orb4, color='purple')
plt.plot(num, orb5, color='green')
plt.hlines(0, num[0], num[-1], ls='--', color='grey')

plt.xticks([0, size1, size2, size3], [r'$\Gamma$', r'$X$', r'$M$', r'$\Gamma$'])
plt.show()
```

```python
dos_data = np.loadtxt('DOS.dat')
w = dos_data[:,0]
dos = dos_data[:,1]
#dos = dos/max(dos)
```

```python
Y_detail = w[4000: 5000]
X_detail = dos[4000: 5000]/max(dos)
```

```python
plt.plot(dos, w)
#plt.xlim(0,0.006)
```

```python
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(10, 5))
ax[0].plot(num, orb1, color='blue')
ax[0].plot(num, orb2, color='orangered')
ax[0].plot(num, orb3, color='gold')
ax[0].plot(num, orb4, color='purple')
ax[0].plot(num, orb5, color='green')
ax[0].hlines(0, num[0], num[-1], ls='--', color='grey')
ax[0].set_xticks([0, size1, size2, size3], [r'$\Gamma$', r'$X$', r'$M$', r'$\Gamma$'])
ax[0].set_ylabel(r'$E-E_F$(eV)')
ax[0].set_ylim(-2,1.75)

ax[1].plot(dos, w)
ax[1].hlines(0,0,max(dos), ls='--', color='grey')
ax[1].set_xlim(0,max(dos))
ax[1].set_xticks([])
ax[1].set_xlabel(r'$DOS$')

sub_axes = plt.axes([.77, .2, .20, .25], xticklabels=[]) 
sub_axes.plot(X_detail, Y_detail) 

plt.tight_layout()
plt.savefig('band.pdf')
plt.show()
```
