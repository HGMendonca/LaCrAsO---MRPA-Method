```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
data1c = np.loadtxt('../gap/res/vecar64sgn035a-Cdk.dat')
size=30
fig, ax=plt.subplots()

mima = np.max((np.abs(data1c[0:28,2].min()), np.abs(data1c[0:28,2].max())))
norm = colors.Normalize(-mima, +mima)
new1 = ax.scatter(x=data1c[0:28,0], y=data1c[0:28,1], c=data1c[0:28,2],s=size, cmap="seismic", norm=norm)  
ax.scatter(x=data1c[0:28,0], y=data1c[0:28,1], s=size, facecolors='none', edgecolors='black', alpha=0.5)
ax.set_ylabel(r'$k_y$') 
ax.set_xlabel(r'$k_x$') 
#ax[0].annotate('$U=0.53;\lambda=0.91$', xy = (-np.pi, np.pi-0.1))
#ax[0].title.set_text('Without $dk$')
fig.colorbar(new1)

plt.savefig('g-eigenv-64x64-Cdk.pdf')
```

```python
data1s = np.loadtxt('../gap/res/vecar64sgn035a-Sdk.dat')
size=30
fig, ax=plt.subplots()

mima = np.max((np.abs(data1s[:,2].min()), np.abs(data1s[:,2].max())))
norm = colors.Normalize(-mima, +mima)
new1 = ax.scatter(x=data1s[:,0], y=data1s[:,1], c=data1s[:,2],s=size, cmap="seismic", norm=norm)  
ax.scatter(x=data1s[:,0], y=data1s[:,1], s=size, facecolors='none', edgecolors='black', alpha=0.5)
ax.set_ylabel(r'$k_y$') 
ax.set_xlabel(r'$k_x$') 
#ax[0].annotate('$U=0.53;\lambda=0.91$', xy = (-np.pi, np.pi-0.1))
#ax[0].title.set_text('Without $dk$')
fig.colorbar(new1)

plt.savefig('g-eigenv-64x64-Sdk.pdf')
```

```python
fig, ax=plt.subplots(1,2, sharey=True, figsize=(11,4))

mima = np.max((np.abs(data1c[:,2].min()), np.abs(data1c[:,2].max())))
norm = colors.Normalize(-mima, +mima)
new1 = ax[0].scatter(x=data1c[:,0], y=data1c[:,1], c=data1c[:,2],s=size, cmap="seismic", norm=norm)  
ax[0].scatter(x=data1c[:,0], y=data1c[:,1], s=size, facecolors='none', edgecolors='black', alpha=0.5)
ax[0].set_ylabel(r'$k_y$') 
ax[0].set_xlabel(r'$k_x$') 
ax[0].annotate('$U=0.60;\lambda=0.4849124$', xy = (-np.pi, np.pi-0.1))
ax[0].title.set_text('With $dk$')
fig.colorbar(new1)

mima = np.max((np.abs(data1s[:,2].min()), np.abs(data1s[:,2].max())))
norm = colors.Normalize(-mima, +mima)
new1 = ax[1].scatter(x=data1s[:,0], y=data1s[:,1], c=data1s[:,2],s=size, cmap="seismic", norm=norm)  
ax[1].scatter(x=data1s[:,0], y=data1s[:,1], s=size, facecolors='none', edgecolors='black', alpha=0.5)
ax[1].set_xlabel(r'$k_x$') 
ax[1].annotate('$U=0.51;\lambda=0.4522464$', xy = (-np.pi, np.pi-0.1))
ax[1].title.set_text('Without $dk$')
fig.colorbar(new1)

plt.savefig('g64x64-comp.pdf')
plt.show()
```
